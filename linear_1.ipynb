### Linear Regression With One Variable

```python
import pandas as pd
import numpy as np
```

```python
df = pd.DataFrame({'area':(2600,3000,3200,3600,4000),
                   'price':(550000,565000,610000,680000,725000)})
df
```

```python
df.info()
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
plt.xlabel('area', size=12)
plt.ylabel('price', size=12)
plt.scatter(df.area,df.price,marker='*')
```

```python
X = df[['area']]  
X
```

```python
y = df.price
y
```

```python
# import our module
from sklearn.linear_model import LinearRegression
```

```python
# Create linear regression object
reg = LinearRegression()
reg.fit(X,y)
```

#### (1) Predict price of a home with area = 3200 sqr ft

```python
reg.predict([[3200]])
```

```python
reg.coef_
```

```python
reg.intercept_
```

### Y = m * X + b (m is coefficient and b is intercept)

```python
3200*135.78767123 + 180616.43835616432
```

#### (2) Predict price of a home with area = 5000 sqr ft

```python
reg.predict([[4000]])
```

### R-Squared value

```python
y
```

```python
df
```

```python
y_predicted=reg.predict(X)
```

```python
y_predicted
```

```python
from sklearn.metrics import r2_score
r2_score(y,y_predicted)
```

### Mean Squared Error 

```python
from sklearn.metrics import mean_squared_error
mean_squared_error(y,y_pred=y_predicted)
```
